Some Advance Correlation.
PEARSON CORRELATION - Measure the strength of the correlation between two features.
1. correlation coefficient
2. P- value.


In our project we are finding the pearson correlation between the two features of the dataset(horsepower and price).

```python
# Pearson correalation between horsepower and price.
import pandas as pd
data = {'horsepower': [100, 150, 200],'curbwieght':[],'' 'price': [10000, 20000, 30000]}
df = pd.DataFrame(data)
from scipy import stats
pearson_coef, p_value = stats.pearsonr(df["horsepower"],df["price"])
pearson_coef,p_value
```

MODEL DEVELOPMENT:
A Model can be thought of as a mathematical equation used to predict a value given one or more other values.

Fitting a simple Linear Model Estimator:

X : Predictor Variable
Y : Taret Variable


```python
# Importing linear_model from scikit-learn
import sklearn as sk
from sklearn.linear_model import LinearRegression

# Creating a Linear Regression Object using the constructor
lm = LinearRegression()

# Now We define the predictor variable and target variable
X = df[["horsepower"]]
Y = df[["price"]]

# Now by using object of LinearRegression class "lm" use fit() method
# to fit the model , i.e. fine the parameter X and Y.
lm.fit(X,Y)

# Now we can obtain the prediction.
yhat=lm.predict(X)
yhat
```

```python
# View the intercept
lm.intercept_
```

```python
# View the slope
lm.coef_
```

Simple Linear Relationship - Estimated Linear Model:


The Relationship between Price and Horsepower is given by:

Price = 10000.0-200*horsepower

MULTIPLE LINEAR REGRESSION:
This method is used to explain the relation between:
1. One continues Target (Y) variable
2. Two or more Precictor (X) variable

Fitting a Multiple Linear Regression Model Estimator:

1. We can extract the 4 predictor variable and stored them in the variable Z.

2. Then train the model as before using fit() method.

3. Obtain the prediction.

```python
# Storing the predictor variable for MLR Model Estimator.
Z = df[["horsepower"]]

# Now training the model
lm.fit(Z,df["price"])
lm.score(Z,df["price"])
```

MODEL EVALUATION USING VISUALIZATION:
